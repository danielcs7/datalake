```python
import os
import duckdb
from minio import Minio
from minio.error import S3Error

# Configuração do MinIO
minio_client = Minio(
    "127.0.0.1:9001",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Buckets e estrutura de pastas no MinIO
source_bucket = "landzone"
dest_bucket = "datalake"
base_prefix = "bronze"

sub_prefix_mapping = {
    "db1": "bronzedb1",
    "db2": "bronzedb2",
    "db3": "bronzedb3"
}

# Criar pastas locais para downloads e exportação
landzone_path = "./downloads_landzone"
bronze_path = "./downloads_bronze"
parquet_export_path = "./parquet_export"

os.makedirs(landzone_path, exist_ok=True)
os.makedirs(bronze_path, exist_ok=True)
os.makedirs(parquet_export_path, exist_ok=True)

# Lista de bancos DuckDB
databases = ["db1", "db2", "db3"]

### 1. BAIXAR ARQUIVOS DA LANDZONE E IMPORTAR PARA O DUCKDB ###
objects = minio_client.list_objects(source_bucket, recursive=True)

for obj in objects:
    file_path = obj.object_name
    file_name = os.path.basename(file_path)

    # Identificar o banco correspondente
    for db, bronze_prefix in sub_prefix_mapping.items():
        if f"_{db}." in file_name:  # Exemplo: fornecedores_db1.parquet
            db_path = f"./{db}.duckdb"
            local_file = os.path.join(landzone_path, file_name)

            # Baixar o arquivo
            minio_client.fget_object(source_bucket, file_path, local_file)
            print(f"Baixado da Landzone: {local_file}")

            # Conectar ao banco e importar
            conn = duckdb.connect(db_path)
            table_name = file_name.replace(f"_{db}.parquet", "")
            conn.execute(f"CREATE TABLE IF NOT EXISTS {table_name} AS SELECT * FROM read_parquet('{local_file}')")
            print(f"Tabela {table_name} criada no banco {db}")

            conn.close()
            os.remove(local_file)  # Excluir o arquivo baixado
            print(f"Arquivo removido: {local_file}")

print("Importação da Landzone concluída!")

### 2. BAIXAR ARQUIVOS DA BRONZE E IMPORTAR PARA O MESMO DUCKDB ###
objects = minio_client.list_objects(dest_bucket, prefix="bronze/", recursive=True)

for obj in objects:
    file_path = obj.object_name
    file_name = os.path.basename(file_path)

    # Identificar o banco correspondente
    for bronze_prefix, db in sub_prefix_mapping.items():
        if file_path.startswith(f"bronze/{bronze_prefix}/"):
            db_path = f"./{db}.duckdb"
            local_file = os.path.join(bronze_path, file_name)

            # Baixar o arquivo
            minio_client.fget_object(dest_bucket, file_path, local_file)
            print(f"Baixado da Bronze: {local_file}")

            # Conectar ao banco e importar
            conn = duckdb.connect(db_path)
            table_name = file_name.replace(".parquet", "")
            conn.execute(f"CREATE TABLE IF NOT EXISTS {table_name} AS SELECT * FROM read_parquet('{local_file}')")
            print(f"Tabela {table_name} criada no banco {db}")

            conn.close()
            os.remove(local_file)  # Excluir o arquivo baixado
            print(f"Arquivo removido: {local_file}")

print("Importação da Bronze concluída!")

### 3. EXPORTAR DADOS DO DUCKDB PARA PARQUET E ENVIAR AO MINIO ###
for db in databases:
    db_path = f"./{db}.duckdb"
    conn = duckdb.connect(db_path)

    # Obter todas as tabelas
    tables = conn.execute("SHOW TABLES").fetchall()

    for table in tables:
        table_name = table[0]
        parquet_file = os.path.join(parquet_export_path, f"{table_name}_{db}.parquet")

        # Exportar dados para Parquet
        conn.execute(f"COPY {table_name} TO '{parquet_file}' (FORMAT PARQUET);")
        print(f"Exportado: {parquet_file}")

        # Caminho no MinIO
        minio_path = f"{base_prefix}/{sub_prefix_mapping[db]}/{table_name}.parquet"

        # Enviar para MinIO
        try:
            minio_client.fput_object(dest_bucket, minio_path, parquet_file)
            print(f"Enviado para MinIO: {dest_bucket}/{minio_path}")

            # Excluir arquivo Parquet local após o upload
            os.remove(parquet_file)
            print(f"Arquivo local removido: {parquet_file}")

        except S3Error as e:
            print(f"Erro ao enviar {parquet_file} para MinIO: {e}")

    conn.close()

# Remover pastas temporárias
try:
    os.rmdir(parquet_export_path)
    os.rmdir(landzone_path)
    os.rmdir(bronze_path)
    print("Pastas temporárias removidas!")
except OSError:
    print("Pastas não estavam vazias ou já foram removidas.")

print("Processo completo! 🚀")
```

```python
from minio import Minio
import duckdb
import os

# Configuração do MinIO
minio_client = Minio(
    "127.0.0.1:9001",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nome dos buckets e estrutura das camadas
buckets = {
    "landzone": ["db1", "db2", "db3"],  # landzone tem arquivos diretamente
    "datalake": ["bronzedb1", "bronzedb2", "bronzedb3"]  # bronze está dentro do datalake
}

# Prefixo para os arquivos bronze dentro do datalake
bronze_prefix = "bronze/"

# Diretório temporário para armazenar os arquivos baixados
download_dir = "minio_downloads"
os.makedirs(download_dir, exist_ok=True)

# Conectar ao DuckDB e criar os schemas
db_path = "dados.duckdb"
conn = duckdb.connect(db_path)

# Criar os schemas necessários
for schemas in buckets.values():
    for schema in schemas:
        conn.execute(f"CREATE SCHEMA IF NOT EXISTS {schema}")

# Função para processar os arquivos de um bucket
def process_bucket(bucket_name, schemas, prefix=""):
    print(f"\n🔍 Listando arquivos no bucket '{bucket_name}' com prefixo '{prefix}'...")
    
    objects = list(minio_client.list_objects(bucket_name, prefix=prefix, recursive=True))
    
    if not objects:
        print(f"⚠️ Nenhum arquivo encontrado no bucket {bucket_name} com prefixo {prefix}!")
        return
    
    for obj in objects:
        file_name = obj.object_name  # Exemplo: "bronze/bronzedb1/fornecedores.parquet"
        print(f"📂 Encontrado: {file_name}")

        if file_name.endswith(".parquet"):
            # Criar a estrutura de diretórios localmente
            local_file = os.path.join(download_dir, os.path.basename(file_name))
            os.makedirs(os.path.dirname(local_file), exist_ok=True)

            # Baixar o arquivo do MinIO
            minio_client.fget_object(bucket_name, file_name, local_file)
            print(f"✅ Arquivo {file_name} baixado para {local_file}")

            # Determinar o schema correto pelo caminho da pasta (não pelo nome do arquivo)
            path_parts = file_name.split("/")
            if len(path_parts) > 2 and path_parts[1] in schemas:
                schema = path_parts[1]  # Exemplo: "bronzedb1"
            else:
                print(f"⚠️ Arquivo {file_name} ignorado (sem correspondência de schema).")
                continue

            # Nome da tabela sem extensão
            table_name = os.path.basename(file_name).replace(".parquet", "").replace("-", "_")
            print(f"📊 Criando tabela: {schema}.{table_name}")

            # Criar a tabela e importar os dados
            query = f"CREATE OR REPLACE TABLE {schema}.{table_name} AS SELECT * FROM read_parquet(?)"
            conn.execute(query, [local_file])  # Passando o caminho do arquivo como parâmetro
            print(f"✅ Arquivo {file_name} carregado no schema {schema} como {table_name}")

# Processar o bucket landzone normalmente
process_bucket("landzone", buckets["landzone"])

# Processar a camada bronze dentro do bucket datalake
process_bucket("datalake", buckets["datalake"], prefix=bronze_prefix)

# Fechar a conexão com o DuckDB
conn.close()
print("\n🎉 Processo concluído!")
```

```python
import duckdb
import pandas as pd
from minio import Minio
import os

# Configuração do MinIO
minio_client = Minio(
    "127.0.0.1:9001",  # URL do MinIO
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Buckets e caminhos das camadas
bucket = "datalake"
bucketZone = "landzone"
landzone_prefix = "landzone/"
bronze_prefix = "bronze/"

# Conectar ao DuckDB
conn = duckdb.connect()

# Configuração do MinIO no DuckDB para leitura S3
conn.execute("SET s3_access_key_id='minioadmin';")
conn.execute("SET s3_secret_access_key='minioadmin';")
conn.execute("SET s3_endpoint='127.0.0.1:9001';")
conn.execute("SET s3_use_ssl=false;")

# Listar arquivos da camada Landzone
objects = minio_client.list_objects(bucketZone, prefix=landzone_prefix, recursive=True)
arquivos_landzone = [obj.object_name for obj in objects if obj.object_name.endswith(".parquet")]

# Exibir os arquivos encontrados na landzone
print("\n📂 Arquivos encontrados na landzone:")
print("\n".join(arquivos_landzone) if arquivos_landzone else "⚠️ Nenhum arquivo encontrado!")

if not arquivos_landzone:
    exit("🚨 ERRO: Nenhum arquivo Parquet encontrado na camada landzone!")

# Percorrer cada arquivo na landzone
for arquivo in arquivos_landzone:
    nome_tabela = arquivo.split("/")[-1]  # Nome do arquivo
    bronze_path = bronze_prefix + nome_tabela  # Caminho na Bronze

    print(f"\n🔍 Processando arquivo: {arquivo}")
    
    # Ler os dados da Landzone com a sintaxe correta
    try:
        df_landzone = conn.execute(
            f"SELECT * FROM read_parquet('s3://{bucket}/{arquivo}')"
        ).fetchdf()

        print(f"✅ Dados carregados de {arquivo} ({len(df_landzone)} linhas).")
        print(df_landzone.head())
        
        if df_landzone.empty:
            print(f"⚠️ O arquivo {arquivo} está vazio! Pulando para o próximo.")
            continue

    except Exception as e:
        print(f"❌ ERRO ao ler {arquivo} da landzone: {e}")
        continue

    # Verificar se o arquivo já existe na Bronze
    try:
        minio_client.stat_object(bucket, bronze_path)
        bronze_existe = True
        print(f"📦 Arquivo {bronze_path} já existe na Bronze.")
    except Exception:
        bronze_existe = False
        print(f"⚠️ Arquivo {bronze_path} NÃO encontrado na Bronze. Será criado.")

    # Definir chave primária (ajuste conforme necessário)
    primary_keys = ["id"]  # Modifique conforme o esquema das suas tabelas

    if bronze_existe:
        # Ler os dados da Bronze
        try:
            df_bronze = conn.execute(
                f"SELECT * FROM read_parquet('s3://{bucket}/{bronze_path}')"
            ).fetchdf()

            print(f"📊 Dados carregados da Bronze ({len(df_bronze)} linhas).")

        except Exception as e:
            print(f"❌ ERRO ao ler {bronze_path}: {e}")
            continue

        # Criar DataFrames indexados para comparação
        lz = df_landzone.set_index(primary_keys)
        bz = df_bronze.set_index(primary_keys)

        # Identificar novas linhas
        novas_linhas = lz.loc[~lz.index.isin(bz.index)].reset_index()

        # Identificar linhas alteradas
        dados_diferentes = lz.loc[lz.index.isin(bz.index) & (lz != bz).any(axis=1)].reset_index()

        print(f"📌 Novas linhas: {len(novas_linhas)} | Linhas alteradas: {len(dados_diferentes)}")

        if not novas_linhas.empty or not dados_diferentes.empty:
            # Criar novo dataframe combinando dados novos e atualizados
            final_df = duckdb.sql("SELECT * FROM df_bronze EXCEPT SELECT * FROM df_landzone")
            final_df = final_df.union(novas_linhas).union(dados_diferentes)

            print(f"📥 Atualizando Bronze com {len(final_df)} registros...")

            # Salvar o arquivo temporário antes de enviar ao MinIO
            temp_path = f"/tmp/{nome_tabela}"
            final_df.to_parquet(temp_path)
            print(f"📥 Arquivo temporário salvo em {temp_path}")

            # Enviar para o MinIO
            with open(temp_path, "rb") as file_data:
                try:
                    minio_client.put_object(bucket, bronze_path, file_data, length=-1, part_size=10*1024*1024)
                    print(f"🚀 Bronze atualizado: {bronze_path}")
                except Exception as e:
                    print(f"❌ Erro ao enviar {bronze_path} para MinIO: {e}")

    else:
        # Criar nova tabela na Bronze
        temp_path = f"/tmp/{nome_tabela}"
        print(f"🆕 Criando nova tabela Bronze: {bronze_path}")

        df_landzone.to_parquet(temp_path)
        print(f"📥 Arquivo temporário salvo em {temp_path}")

        # Enviar para o MinIO
        with open(temp_path, "rb") as file_data:
            try:
                minio_client.put_object(bucket, bronze_path, file_data, length=-1, part_size=10*1024*1024)
                print(f"✅ Tabela {bronze_path} criada na Bronze!")
            except Exception as e:
                print(f"❌ Erro ao criar {bronze_path} na Bronze: {e}")

print("\n✅ Processo concluído!")
```

```python
from minio import Minio
from minio.error import S3Error
import duckdb
import os

# Configuração do cliente MinIO
minio_client = Minio(
    "127.0.0.1:9001",  # URL do MinIO
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nomes dos buckets
landzone_bucket = "landzone"
bronze_bucket = "datalake"

# Função para baixar um arquivo do MinIO para o sistema local
def download_file(bucket_name, file_path, local_path):
    try:
        minio_client.fget_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{file_path}' baixado para '{local_path}'.")
    except S3Error as e:
        print(f"Erro ao baixar o arquivo '{file_path}': {e}")

# Função para fazer upload de um arquivo para o MinIO
def upload_file(bucket_name, file_path, local_path):
    try:
        minio_client.fput_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{local_path}' enviado para '{bucket_name}/{file_path}'.")
    except S3Error as e:
        print(f"Erro ao enviar o arquivo '{local_path}': {e}")

# Função para listar arquivos em um bucket
def list_files(bucket_name, prefix=""):
    try:
        objects = minio_client.list_objects(bucket_name, prefix=prefix, recursive=True)
        return [obj.object_name for obj in objects]
    except S3Error as e:
        print(f"Erro ao listar arquivos no bucket '{bucket_name}': {e}")
        return []

# Função para realizar o upsert usando DuckDB
def upsert_parquet(landzone_file, bronze_file, output_file):
    # Conectar ao DuckDB
    conn = duckdb.connect()

    # Ler os arquivos Parquet
    landzone_df = conn.execute(f"SELECT * FROM read_parquet('{landzone_file}')").fetchdf()
    bronze_df = conn.execute(f"SELECT * FROM read_parquet('{bronze_file}')").fetchdf()

    # Supondo que há uma coluna única para identificar registros (ex: id)
    primary_key = "id"  # Altere para a coluna correta

    # Realizar o upsert
    upsert_query = f"""
    SELECT * FROM landzone_df
    UNION ALL
    SELECT * FROM bronze_df
    WHERE {primary_key} NOT IN (SELECT {primary_key} FROM landzone_df)
    """
    result_df = conn.execute(upsert_query).fetchdf()

    # Salvar o resultado em um novo arquivo Parquet
    result_df.to_parquet(output_file)
    print(f"Upsert concluído. Resultado salvo em '{output_file}'.")

    # Fechar a conexão
    conn.close()

# Verifica se o bucket 'datalake' existe ou cria ele
if not minio_client.bucket_exists(bronze_bucket):
    minio_client.make_bucket(bronze_bucket)
    print(f"Bucket '{bronze_bucket}' criado com sucesso.")

# Lista todos os arquivos na 'landzone'
landzone_files = list_files(landzone_bucket)

# Processa cada arquivo na 'landzone'
for file_path in landzone_files:
    # Extrai o nome do arquivo (ex: fornecedores_db1.parquet)
    file_name = file_path.split("/")[-1]

    # Verifica se o arquivo segue o padrão esperado (ex: tipo_dbX.parquet)
    if "_db" in file_name and file_name.endswith(".parquet"):
        # Extrai o tipo do arquivo (ex: fornecedores, clientes, produtos)
        file_type = file_name.split("_db")[0]

        # Extrai o número do banco (ex: db1, db2, db3)
        db_number = file_name.split("_")[-1].split(".")[0]  # Ex: db1

        # Define o caminho de destino no bucket 'datalake/bronze'
        destination_path = f"bronze/{db_number}/{file_type}/{file_name}"

        # Baixa os arquivos da 'landzone' e da 'bronze' para o sistema local
        local_landzone_file = f"temp_landzone_{file_name}"
        local_bronze_file = f"temp_bronze_{file_name}"
        local_output_file = f"temp_output_{file_name}"

        download_file(landzone_bucket, file_path, local_landzone_file)
        if minio_client.stat_object(bronze_bucket, destination_path):
            download_file(bronze_bucket, destination_path, local_bronze_file)
        else:
            local_bronze_file = None  # Arquivo não existe na 'bronze'

        # Realiza o upsert
        if local_bronze_file:
            upsert_parquet(local_landzone_file, local_bronze_file, local_output_file)
        else:
            # Se o arquivo não existir na 'bronze', apenas copia o arquivo da 'landzone'
            print(f"Arquivo '{file_name}' não existe na 'bronze'. Inserindo...")
            upload_file(bronze_bucket, destination_path, local_landzone_file)

        # Remove os arquivos temporários
        os.remove(local_landzone_file)
        if local_bronze_file:
            os.remove(local_bronze_file)
        if os.path.exists(local_output_file):
            os.remove(local_output_file)
```

```python
from minio import Minio
from minio.error import S3Error
import duckdb
import os

# Configuração do cliente MinIO
minio_client = Minio(
    "127.0.0.1:9001",  # URL do MinIO
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nomes dos buckets
landzone_bucket = "landzone"
bronze_bucket = "datalake"

# Função para baixar um arquivo do MinIO para o sistema local
def download_file(bucket_name, file_path, local_path):
    try:
        minio_client.fget_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{file_path}' baixado para '{local_path}'.")
    except S3Error as e:
        print(f"Erro ao baixar o arquivo '{file_path}': {e}")

# Função para fazer upload de um arquivo para o MinIO
def upload_file(bucket_name, file_path, local_path):
    try:
        minio_client.fput_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{local_path}' enviado para '{bucket_name}/{file_path}'.")
    except S3Error as e:
        print(f"Erro ao enviar o arquivo '{local_path}': {e}")

# Função para listar arquivos em um bucket
def list_files(bucket_name, prefix=""):
    if not minio_client.bucket_exists(bucket_name):
        print(f"Bucket '{bucket_name}' não existe.")
        return []
    
    try:
        objects = minio_client.list_objects(bucket_name, prefix=prefix, recursive=True)
        return [obj.object_name for obj in objects]
    except S3Error as e:
        print(f"Erro ao listar arquivos no bucket '{bucket_name}': {e}")
        return []

# Função para realizar o upsert usando DuckDB
def upsert_parquet(landzone_file, bronze_file, output_file):
    conn = duckdb.connect()

    # Verifica se o arquivo bronze existe
    bronze_exists = os.path.exists(bronze_file)

    # Lê os arquivos Parquet
    landzone_df = conn.execute(f"SELECT * FROM read_parquet('{landzone_file}')").fetchdf()
    if bronze_exists:
        bronze_df = conn.execute(f"SELECT * FROM read_parquet('{bronze_file}')").fetchdf()
    else:
        bronze_df = None

    # Supondo que há uma coluna única para identificar registros (ex: id)
    primary_key = "id"  # Altere para a coluna correta

    if bronze_df is not None:
        upsert_query = f"""
        SELECT * FROM landzone_df
        UNION ALL
        SELECT * FROM bronze_df
        WHERE {primary_key} NOT IN (SELECT {primary_key} FROM landzone_df)
        """
        result_df = conn.execute(upsert_query).fetchdf()
    else:
        result_df = landzone_df  # Apenas insere os novos dados se não houver bronze

    # Salva o resultado em um novo arquivo Parquet
    result_df.to_parquet(output_file)
    print(f"Upsert concluído. Resultado salvo em '{output_file}'.")

    conn.close()

# Verifica se o bucket 'datalake' existe ou cria ele
if not minio_client.bucket_exists(bronze_bucket):
    minio_client.make_bucket(bronze_bucket)
    print(f"Bucket '{bronze_bucket}' criado com sucesso.")

# Lista todos os arquivos na 'landzone'
landzone_files = list_files(landzone_bucket)

# Processa cada arquivo na 'landzone'
for file_path in landzone_files:
    file_name = file_path.split("/")[-1]

    # Verifica se o arquivo segue o padrão esperado (ex: tipo_dbX.parquet)
    if "_db" in file_name and file_name.endswith(".parquet"):
        file_type = file_name.split("_db")[0]
        db_number = file_name.split("_")[-1].split(".")[0]  # Ex: db1

        # Caminho no bucket 'bronze' (correção aqui)
        destination_path = f"bronze/{db_number}/{file_name}"

        # Arquivos locais temporários
        local_landzone_file = f"temp_landzone_{file_name}"
        local_bronze_file = f"temp_bronze_{file_name}"
        local_output_file = f"temp_output_{file_name}"

        # Baixa o arquivo da 'landzone'
        download_file(landzone_bucket, file_path, local_landzone_file)

        # Verifica se o arquivo existe na 'bronze' antes de tentar baixá-lo
        try:
            minio_client.stat_object(bronze_bucket, destination_path)
            download_file(bronze_bucket, destination_path, local_bronze_file)
            bronze_exists = True
        except S3Error:
            print(f"Arquivo '{destination_path}' não encontrado na bronze. Será inserido diretamente.")
            bronze_exists = False

        # Realiza o upsert
        try:
            if bronze_exists:
                upsert_parquet(local_landzone_file, local_bronze_file, local_output_file)
                upload_file(bronze_bucket, destination_path, local_output_file)
            else:
                upload_file(bronze_bucket, destination_path, local_landzone_file)
        finally:
            # Remove arquivos temporários
            for temp_file in [local_landzone_file, local_bronze_file, local_output_file]:
                if os.path.exists(temp_file):
                    os.remove(temp_file)
```

## CAMADA SILVERANALITICS

```python
import duckdb
import pandas as pd
from minio import Minio
import os

# Configuração do cliente MinIO
minio_client = Minio(
    "127.0.0.1:9001",  # URL do MinIO
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nomes dos buckets
datalake_bucket = "datalake"  # O bucket 'datalake' sem a barra
silver_prefix = "silver/"  # Prefixo para arquivos em 'silver'
silveranalitics_prefix = "silveranalitics/"  # Prefixo para arquivos em 'silveranalitics'

# Função para baixar um arquivo do MinIO para o sistema local
def download_file(bucket_name, file_path, local_path):
    try:
        minio_client.fget_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{file_path}' baixado para '{local_path}'.")
    except Exception as e:
        print(f"Erro ao baixar o arquivo '{file_path}': {e}")

# Função para fazer upload de um arquivo para o MinIO
def upload_file(bucket_name, file_path, local_path):
    try:
        minio_client.fput_object(bucket_name, file_path, local_path)
        print(f"Arquivo '{local_path}' enviado para '{bucket_name}/{file_path}'.")
    except Exception as e:
        print(f"Erro ao enviar o arquivo '{local_path}': {e}")

# Função para listar arquivos em um bucket
def list_files(bucket_name, prefix=""):
    if not minio_client.bucket_exists(bucket_name):
        print(f"Bucket '{bucket_name}' não existe.")
        return []
    
    try:
        objects = minio_client.list_objects(bucket_name, prefix=prefix, recursive=True)
        return [obj.object_name for obj in objects]
    except Exception as e:
        print(f"Erro ao listar arquivos no bucket '{bucket_name}': {e}")
        return []

# Função para fazer o upsert dos dados para o bucket silveranalitics
def upsert_to_silveranalitics():
    # Conectando ao DuckDB em memória
    con = duckdb.connect()

    # Baixar todos os arquivos na camada silver
    db_files = list_files(datalake_bucket, silver_prefix + "db1/")
    db_files += list_files(datalake_bucket, silver_prefix + "db2/")
    db_files += list_files(datalake_bucket, silver_prefix + "db3/")  # Adicione mais dbs se necessário

    # Agrupar os arquivos por tipo (fornecedores, produtos, etc.)
    file_types = set([f.split("/")[-1].split("_")[0] for f in db_files])  # Ex: fornecedores, produtos, etc.

    for file_type in file_types:
        # Filtrar os arquivos de acordo com o tipo
        type_files = [f for f in db_files if file_type in f and f.endswith(".parquet")]

        # DataFrame vazio para juntar os dados de cada tipo
        type_df = pd.DataFrame()

        for f in type_files:
            # Baixar e carregar os arquivos
            local_file = f"temp_{f.split('/')[-1]}"
            download_file(datalake_bucket, f, local_file)

            # Ler o arquivo .parquet
            temp_df = pd.read_parquet(local_file)
            type_df = pd.concat([type_df, temp_df], ignore_index=True)

            # Remover arquivo temporário
            if os.path.exists(local_file):
                os.remove(local_file)

        # Realizar o upsert (aqui estamos fazendo um exemplo de concatenação e remoção de duplicatas)
        type_df = type_df.drop_duplicates()

        # Definir o caminho do arquivo na camada silveranalitics
        silveranalitics_file = silveranalitics_prefix + f"{file_type}.parquet"

        # Salvar o arquivo atualizado na camada silveranalitics
        type_df.to_parquet(f"temp_{file_type}.parquet")

        # Fazer o upload para o MinIO
        upload_file(datalake_bucket, silveranalitics_file, f"temp_{file_type}.parquet")

        # Remover arquivos temporários locais
        if os.path.exists(f"temp_{file_type}.parquet"):
            os.remove(f"temp_{file_type}.parquet")

# Verifica se o bucket 'datalake' existe ou cria ele
if not minio_client.bucket_exists(datalake_bucket):
    minio_client.make_bucket(datalake_bucket)
    print(f"Bucket '{datalake_bucket}' criado com sucesso.")

# Verifica se o bucket 'silveranalitics' existe ou cria ele
if not minio_client.bucket_exists(silveranalitics_prefix):
    minio_client.make_bucket(silveranalitics_prefix)
    print(f"Bucket '{silveranalitics_prefix}' criado com sucesso.")

# Processa todos os arquivos da camada silver e faz o upsert na silveranalitics
upsert_to_silveranalitics()
```
