# TESTES DE PY VIGENTE

```python
import duckdb
import os
import glob

# Caminho onde os arquivos Parquet foram baixados
local_download_path = "./downloads_landzone/landzone"

# Identifica todos os arquivos Parquet
parquet_files = glob.glob(os.path.join(local_download_path, "*.parquet"))

# Dicionário para armazenar conexões DuckDB
db_connections = {}

# Processar cada arquivo Parquet
for file in parquet_files:
    # Extrair nome do arquivo (ex: fornecedores_db1.parquet)
    filename = os.path.basename(file)
    
    # Separar nome base e banco (ex: "fornecedores", "db1")
    table_name, db_name = filename.replace(".parquet", "").rsplit("_", 1)

    # Criar conexão com o banco correspondente
    db_path = f"./downloads_landzone/{db_name}.duckdb"
    
    if db_name not in db_connections:
        db_connections[db_name] = duckdb.connect(db_path)

    conn = db_connections[db_name]

    # Criar tabela e importar os dados
    conn.execute(f"""
        CREATE TABLE IF NOT EXISTS {table_name} AS 
        SELECT * FROM read_parquet('{file}');
    """)
    print(f"Arquivo {filename} importado para {db_name}.duckdb na tabela {table_name}.")

# Fechar conexões
for conn in db_connections.values():
    conn.close()

print("Importação concluída!")
```

```python
import duckdb
import os
from minio import Minio
from minio.error import S3Error

# Configuração do MinIO
minio_client = Minio(
    "127.0.0.1:9001",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nome do bucket de destino
dest_bucket = "datalake"

# Estrutura das pastas no MinIO
base_prefix = "bronze"
sub_prefix_mapping = {
    "db1": "bronzedb1",
    "db2": "bronzedb2",
    "db3": "bronzedb3"
}

# Lista de bancos DuckDB
databases = ["db1", "db2", "db3"]

# Criar pasta local temporária para armazenar os Parquets antes do upload
local_parquet_path = "./parquet_export"
os.makedirs(local_parquet_path, exist_ok=True)

# Processar cada banco
for db in databases:
    db_path = f"./downloads_landzone/{db}.duckdb"
    conn = duckdb.connect(db_path)

    # Obter todas as tabelas do banco
    tables = conn.execute("SHOW TABLES").fetchall()
    
    for table in tables:
        table_name = table[0]

        # Caminho do arquivo Parquet
        parquet_file = os.path.join(local_parquet_path, f"{table_name}_{db}.parquet")

        # Exportar os dados para Parquet
        conn.execute(f"COPY {table_name} TO '{parquet_file}' (FORMAT PARQUET);")
        print(f"Exportado: {parquet_file}")

        # Caminho no MinIO
        minio_path = f"{base_prefix}/{sub_prefix_mapping[db]}/{table_name}.parquet"

        # Enviar para MinIO
        try:
            minio_client.fput_object(dest_bucket, minio_path, parquet_file)
            print(f"Enviado para MinIO: {dest_bucket}/{minio_path}")
        except S3Error as e:
            print(f"Erro ao enviar {parquet_file} para MinIO: {e}")

    # Fechar conexão com o banco DuckDB
    conn.close()

print("Processo concluído! 🎯")
```

```python
import duckdb
import os
from minio import Minio
from minio.error import S3Error

# Configuração do MinIO
minio_client = Minio(
    "127.0.0.1:9001",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Nome do bucket de destino
dest_bucket = "datalake"

# Estrutura das pastas no MinIO
base_prefix = "bronze"
sub_prefix_mapping = {
    "db1": "bronzedb1",
    "db2": "bronzedb2",
    "db3": "bronzedb3"
}

# Lista de bancos DuckDB
databases = ["db1", "db2", "db3"]

# Criar pasta local temporária para armazenar os Parquets antes do upload
local_parquet_path = "./parquet_export"
os.makedirs(local_parquet_path, exist_ok=True)

# Processar cada banco
for db in databases:
    db_path = f"./downloads_landzone/{db}.duckdb"
    conn = duckdb.connect(db_path)

    # Obter todas as tabelas do banco
    tables = conn.execute("SHOW TABLES").fetchall()
    
    for table in tables:
        table_name = table[0]

        # Caminho do arquivo Parquet
        parquet_file = os.path.join(local_parquet_path, f"{table_name}_{db}.parquet")

        # Exportar os dados para Parquet
        conn.execute(f"COPY {table_name} TO '{parquet_file}' (FORMAT PARQUET);")
        print(f"Exportado: {parquet_file}")

        # Caminho no MinIO
        minio_path = f"{base_prefix}/{sub_prefix_mapping[db]}/{table_name}.parquet"

        # Enviar para MinIO
        try:
            minio_client.fput_object(dest_bucket, minio_path, parquet_file)
            print(f"Enviado para MinIO: {dest_bucket}/{minio_path}")

            # Excluir arquivo Parquet local após o upload
            os.remove(parquet_file)
            print(f"Arquivo local removido: {parquet_file}")

        except S3Error as e:
            print(f"Erro ao enviar {parquet_file} para MinIO: {e}")

    # Fechar conexão com o banco DuckDB
    conn.close()

# Remover a pasta temporária se estiver vazia
try:
    os.rmdir(local_parquet_path)
    print(f"Pasta removida: {local_parquet_path}")
except OSError:
    print(f"A pasta {local_parquet_path} não está vazia ou já foi removida.")

print("Processo concluído! 🎯")
```
